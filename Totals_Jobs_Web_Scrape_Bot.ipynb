```python
import requests
from bs4 import BeautifulSoup
import lxml
import pandas as pd
```

```python
url_total_jobs = 'https://www.totaljobs.com/jobs/in-uk?cid=SEA_GO_UK-TJ-Gen---PMAX%7C%5BA%5D_c_PMAX_PMAX_PMAX_PMAX001&gad_source=1&gclid=CjwKCAiA3ZC6BhBaEiwAeqfvymWVn6IyAfM_rZuOHGCvU2vvOSUajlsdSfWwybIZAK8BBnC2QdLCJRoChygQAvD_BwE'
response_total_jobs = requests.get(url_total_jobs)
html_content_total_jobs = response_total_jobs.text
```

```python
soup = BeautifulSoup(html_content_total_jobs, 'lxml')

total_jobs_title = soup.find_all(name='div', class_='res-nehv70')
total_jobs_company_name = soup.find_all(name='span', class_='res-btchsq')
total_jobs_salary = soup.find_all(name='span', attrs={'data-at': 'job-item-salary-info'})

total_jobs_title_storage = []
total_jobs_company_name_storage = []
total_jobs_salary_storage = []

for title in total_jobs_title:
  total_jobs_title_storage.append(title.text)

for name in total_jobs_company_name:
  total_jobs_company_name_storage.append(name.text)

for salary in total_jobs_salary:
  total_jobs_salary_storage.append(salary.text)
```
