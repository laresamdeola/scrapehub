```python
import requests
from bs4 import BeautifulSoup
import lxml
import pandas as pd
```

```python
url_dimi = 'https://thebetterindia.com/topics/blog/'
response_dimi = requests.get(url_dimi)
html_content_dimi = response_dimi.text
```

```python
html_content_dimi
```

```python
soup = BeautifulSoup(html_content_dimi, 'lxml')

total_dimi_heading = soup.find_all(name='h3', class_='heading')
# total_dimi_description = soup.find_all(name='div', class_='pt-cv-content')

total_dimi_heading_storage = []
# total_dimi_description_storage = []
```

```python
total_dimi_heading
```

```python
for heading in total_dimi_heading:
  total_dimi_heading_storage.append(heading.text)
```

```python
total_dimi_heading_storage
```

```python
dimi_dataframe = pd.DataFrame(
    {'Title': total_dimi_heading_storage
     })
```

```python
dimi_dataframe
```

```python
dimi_dataframe.to_csv('dimi_dataframe_2.txt', index=False, header=False)
```
