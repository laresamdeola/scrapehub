```python
import requests
from bs4 import BeautifulSoup
import lxml
import pandas as pd
```

```python
url = 'https://play.google.com/store/apps'
url_2 = 'https://play.google.com/store/search?q=education&c=apps'
response = requests.get(url_2)
html_content = response.text
```

```python
html_content
```

```python
soup = BeautifulSoup(html_content, 'lxml')

top_free_apps_names = soup.find_all(name='span', class_='wMUdtb')
top_free_apps_rating = soup.find_all(name='span', class_='w2kbF')
top_free_apps_heading = soup.find_all(name='span', class_='DdYX5')

app_names = []
app_rating = []
app_heading = []

for app in top_free_apps_names:
  app_names.append(''.join(app))

for rate in top_free_apps_rating:
  app_rating.append(float(''.join(rate)))

for head in top_free_apps_heading:
  app_heading.append(''.join(head))
```

```python
google_app = pd.DataFrame({
    'name': app_names,
    'heading': app_heading,
    'rating': app_rating
})
```
